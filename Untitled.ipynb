```python
import cudf
import glob
import os

# Configuration
INPUT_DIR = "processed_parquet"

def load_cs_papers():
    print("üöÄ Scanning for Computer Science papers on GPU (Fixed)...")
    
    files = sorted(glob.glob(f"{INPUT_DIR}/*.parquet"))
    cs_chunks = []
    
    for i, file_path in enumerate(files):
        try:
            # 1. Load the chunk
            gdf = cudf.read_parquet(file_path)
            
            # 2. Define the Filter Condition (The Fix: .str.lower() first)
            # We look for "computer science" in lowercased fields
            # We use fillna(False) because null strings cause GPU errors in boolean masks
            mask = (
                gdf['primary_field'].str.lower().str.contains("computer science", regex=False).fillna(False) | 
                gdf['all_fields'].str.lower().str.contains("computer science", regex=False).fillna(False) | 
                gdf['ext_fields'].str.lower().str.contains("computer science", regex=False).fillna(False)
            )
            
            # 3. Apply Filter
            cs_only = gdf[mask]
            
            # 4. Keep chunk if not empty
            if len(cs_only) > 0:
                cs_chunks.append(cs_only)
                
            if i % 50 == 0:
                print(f"   Processed {i}/{len(files)} files...")
                
        except Exception as e:
            print(f"‚ùå Error in {file_path}: {e}")

    # 5. Concatenate all chunks into one massive GPU DataFrame
    if cs_chunks:
        final_cs_df = cudf.concat(cs_chunks, ignore_index=True)
        print(f"\n‚úÖ LOADED: {len(final_cs_df):,} Computer Science papers.")
        return final_cs_df
    else:
        print("\n‚ùå No Computer Science papers found.")
        return None

# Execute
cs_df = load_cs_papers()

# Preview
if cs_df is not None:
    print("\nüëÄ Preview of CS Papers:")
    print(cs_df[['paper_id', 'year', 'primary_field', 'all_fields']].head())
```
